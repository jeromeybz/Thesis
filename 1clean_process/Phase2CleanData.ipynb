```python
import pandas as pd
import gdown

# Google Drive file ID
file_id = '1dD_YytfGq7JQI_wBPvGnmpRGBgZXict-'

# Generate the download link
url = f'https://drive.google.com/uc?export=download&id={file_id}'

# Download the file
output = 'file.csv'
gdown.download(url, output, quiet=False)

# Try reading the CSV with a different encoding
dataset = pd.read_csv(output, encoding='ISO-8859-1')
dataset.columns = ['Sentiment', 'id', 'date', 'query', 'user', 'Text']

# Show the first few rows of the DataFrame
dataset.head()
```

```python
#drop
df = dataset.copy()
df.drop(dataset.columns[[1, 2, 3, 4]], axis=1, inplace=True)
df.head()
```

```python
import re

def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove mentions (@username) and hashtags (#hashtag)
    text = re.sub(r'\@\w+|\#', '', text)
    # Replace '&' with 'and'
    text = re.sub(r'&', 'and', text)
    # Remove extra punctuation but keep the first occurrence
    text = re.sub(r'([.,!?"])\1+', r'\1', text)  # Replace repeated punctuation with a single instance
    # Remove trailing punctuation (except '!')
    text = re.sub(r'([.,?"]$)', '', text)  # Keep '!'
    # Remove apostrophes but keep other characters
    text = re.sub(r"'", '', text)  # Remove only '
    # Remove numbers
    text = re.sub(r'\d+', '', text)  # Removes all numbers (0-9)
    # Remove unwanted symbols, including semicolons (;), underscores, dashes, and math symbols (+, =)
    text = re.sub(r'[\(\)\%\^\*\+/\{\}\[\]\\|~$€£¥₹;©®™§¶_\-+=]', '', text)  # Remove specified symbols
    # Remove emojis and other special Unicode symbols
    text = re.sub(r'[\U00002000-\U0001FAFF]', '', text)  # Removes most emojis & symbols
    # Remove fractions, superscripts, and subscripts (Unicode ranges)
    text = re.sub(r'[\u00BC-\u00BE\u2150-\u218F]', '', text)  # Fractions (¼ ½ ¾), superscripts, subscripts
    # Remove non-ASCII characters (e.g., â«, â€™, Ã, etc.)
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # Removes all non-ASCII characters
    # Remove leading spaces
    text = re.sub(r'^\s+', '', text)
    # Normalize whitespace (remove extra spaces)
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    # Convert to lowercase
    text = text.lower()

    return text

# Apply the function to the dataset
df['clean_text'] = df['Text'].apply(clean_text)
df
```

```python
#find empty rows
empty_clean_text = df[df['clean_text'].isna() | (df['clean_text'] == '')]

empty_clean_text
```

```python
# delete null rows
df = df[df['clean_text'].notna() & (df['clean_text'] != '')]
df = df.reset_index(drop=True)
print(df.head())
```

```python
df
```

```python
#1 sentence per row
expanded_rows = []

for _, row in df.iterrows():
    sentences = row["clean_text"].split(".")
    sentences = [s.strip() for s in sentences if s.strip()]
    for sentence in sentences:
        expanded_rows.append([sentence, row["Sentiment"]])

expand_Dataset = pd.DataFrame(expanded_rows, columns=["clean_text", "Sentiment"])
expand_Dataset
```

```python
# find empty rows
empty_clean_text = expand_Dataset[expand_Dataset['clean_text'].isna() | (expand_Dataset['clean_text'] == '')]
empty_clean_text
```

```python
# Save the DataFrame to a CSV file
expand_Dataset.to_csv('cleaned_dataset_1row1sentencee.csv', index=False)

# Optional: Provide feedback to confirm the file has been saved
print("CSV file 'cleaned_dataset.csv' has been saved.")
```
