```python
import pandas as pd
import gdown

# Google Drive file ID
file_id = '1afyZ_pQkJSgv83Z2TaxAJzPf_iUNrTKl'

# Generate the download link
url = f'https://drive.google.com/uc?export=download&id={file_id}'

# Download the file
output = 'file.csv'
gdown.download(url, output, quiet=False)

# Try reading the CSV with a different encoding
dataset = pd.read_csv(output, encoding='ISO-8859-1')

# Show the first few rows of the DataFrame
dataset
```

```python
import ast
```

```python
# convert postags into list
dataset["pos_tags"] = dataset["pos_tags"].apply(ast.literal_eval)
```

```python
# finds the max length of pos tags in any row
max_length = dataset["pos_tags"].apply(len).max()
max_length
```

```python

empty_clean_text = dataset[dataset['pos_tags'].isna() | (dataset['pos_tags'] == '')]

empty_clean_text
```

```python
empty_clean_text = dataset[dataset['Sentiment'].isna() | (dataset['Sentiment'] == '')]

empty_clean_text
```

```python
#get max length 
pos_columns = [f"pos_{i+1}" for i in range(max_length)]
expanded_df = pd.DataFrame(dataset["pos_tags"].to_list(), columns=pos_columns)

expanded_df.insert(0, "Sentiment", dataset["Sentiment"])

expanded_df.to_csv("expanded_pos_tags6removed.csv", index=False)
```
